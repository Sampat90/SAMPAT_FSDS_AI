```python
vis2 = sns.displot(df['InternetUsers'])
plt.show(vis2)
```

```python
vis2 = sns.distplot(df['InternetUsers'],bins=15)
plt.show(vis2)
```

```python
vis3 = sns.boxplot(data=df,x="IncomeGroup",y='BirthRate')
plt.show(vis3)  # Bialateral Analysis outlier

# we need probability to remove/adjust the oulier/ANAMOLY DETECTION (sigmoid function)
```

```python
vis4 = sns.lmplot(data=df,x="InternetUsers",y='BirthRate')
plt.show(vis4)
```

```python
vis5 = sns.lmplot(data=df,x="InternetUsers",y='BirthRate',fit_reg='False')
plt.show(vis5)
```

```python
vis6 = sns.lmplot(data=df,x="InternetUsers",y='BirthRate',fit_reg='True')
plt.show(vis6)
```

```python
vis7 = sns.lmplot(data=df,x="InternetUsers",y='BirthRate',fit_reg='False',hue='IncomeGroup') #hue parameter for colour
plt.show(vis7)
```

```python
vis8 = sns.lmplot(data=df,x="InternetUsers",y='BirthRate',fit_reg='True',hue='IncomeGroup') #hue parameter for colour
plt.show(vis8)
```
